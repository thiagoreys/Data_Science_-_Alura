Temos um [arquivo CSV](https://raw.githubusercontent.com/alura-cursos/Pandas/main/dados_sus.csv), que foi obtido na pÃ¡gina InformaÃ§Ãµes de SaÃºde (TABNET) - DATASUS [texto do link](https://datasus.saude.gov.br/informacoes-de-saude-tabnet/). Este arquivo contÃ©m os gastos hospitalares pÃºblicos de cada uma das Unidades da FederaÃ§Ã£o brasileira no perÃ­odo de janeiro de 2008 a marÃ§o de 2021.

Seu desafio aqui serÃ¡ fazer a leitura desse arquivo utilizando a funÃ§Ã£o `read_csv` da biblioteca Pandas. Alguns parÃ¢metros devem ser adicionados para que a leitura seja feita corretamente. EntÃ£o, seguem algumas dicas:

* Verifique se o arquivo CSV estÃ¡ separado por vÃ­rgula ou ponto e vÃ­rgula.
* A codificaÃ§Ã£o do arquivo Ã© ISO-8859-1.
* As trÃªs primeiras linhas linhas do arquivo podem ser desconsideradas, pois o cabeÃ§alho sÃ³ comeÃ§a na quarta linha.
* As 9 Ãºltimas linhas tambÃ©m podem ser desconsideradas, pois sÃ£o apenas informaÃ§Ãµes sobre onde os dados foram obtidos.

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/dados_sus.csv'
```

```python
dados_sus = pd.read_csv(url, encoding='ISO-8859-1', sep=';', skiprows=3, skipfooter=9, engine='python')
```

`engine='python'`: o motor usado para ler o arquivo Ã© o Python. Isso Ã© necessÃ¡rio quando se usa o parÃ¢metro skipfooter, pois o motor padrÃ£o nÃ£o suporta essa opÃ§Ã£o.
