### Tokenização

O processo de tokenização consiste em remover tudo o que não for relevante para analise de dados nas strings, deixando apenas as palavras separadas em listas.

#### Tokenizando a coluna 'descricao_local'

```python
# Removendo tudo o que não for letras, números, hífen (-) e apóstrofo (').
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-z0-9\-\']', ' ', regex=True)
```

* `^` (negação dos argumentos, faz com que se busque tudo o que não for eles)
* `\` (anula caractere especial)



```python
# Removendo hífens isolados
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?<!\w)', ' ', regex=True)
```

```python
# dados['descricao_local'] = dados['descricao_local'].str.replace(' - ', ' ')
```

De maneira geral, o regex permite que os cientistas de dados processem, analisem e classifiquem grandes volumes de dados de texto de forma eficiente e automatizada. O uso adequado do regex pode ajudar a extrair informações valiosas de dados de texto, além de facilitar a limpeza e organização desses dados.

#### **Projeto 2: administração de condomínios**

Assim como no projeto desafio 1, trabalhamos com todas as colunas exceto as que envolvem datas.

Nas colunas `datas_de_pagamento` e `datas_combinadas_pagamento`, temos datas em formato 'dia/mês/ano' (`dd/mm/AAAA`). Transforme esses dados para o tipo datetime e busque uma forma de visualização de subconjunto que possa contribuir no objetivo do contexto que os dados estão inseridos.

Se você não lembra o problema do projeto desafio 2, vou deixar abaixo o texto da situação para facilitar o encontro da informação:

> A administração de condomínios é uma tarefa que requer muita atenção e organização. Entre as diversas responsabilidades da gestão está o recebimento dos aluguéis dos locatários. Para garantir uma boa saúde financeira do empreendimento, é fundamental que esses pagamentos sejam feitos de forma regular e pontual. No entanto, sabemos que nem sempre isso acontece. Pensando nisso, proponho um desafio de tratamento de dados com o objetivo de analisar o atraso no pagamento de aluguéis no condomínio fictício de alguns moradores.

```python
dados_p2.head()
```

```python
dados_p2['datas_combinadas_pagamento'] = pd.to_datetime(dados_p2['datas_combinadas_pagamento'], dayfirst=True)
dados_p2['datas_de_pagamento'] = pd.to_datetime(dados_p2['datas_de_pagamento'], dayfirst=True)
dados_p2
```

```python
# Calculando dias de atraso
dias_atrasados = (dados_p2['datas_de_pagamento'] - dados_p2['datas_combinadas_pagamento']).dt.days
dados_p2.insert(3, 'dias_atrasados', dias_atrasados)
dados_p2.head()
```

```python
atrasos_alugueis = dados_p2.groupby('apartamento').agg(dias_atrasados=('dias_atrasados', 'sum'),
                                              média_atraso=('dias_atrasados', 'mean'),
                                              valor_aluguel=('valor_aluguel', 'last')).sort_values('dias_atrasados')
atrasos_alugueis
```
