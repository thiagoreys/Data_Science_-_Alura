### TokenizaÃ§Ã£o

O processo de tokenizaÃ§Ã£o consiste em remover tudo o que nÃ£o for relevante para analise de dados nas strings, deixando apenas as palavras separadas em listas.

#### Tokenizando a coluna 'descricao_local'

```python
# Removendo tudo o que nÃ£o for letras, nÃºmeros, hÃ­fen (-) e apÃ³strofo (').
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-z0-9\-\']', ' ', regex=True)
```

* `^` (negaÃ§Ã£o dos argumentos, faz com que se busque tudo o que nÃ£o for eles)
* `\` (anula caractere especial)



```python
# Removendo hÃ­fens isolados
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?<!\w)', ' ', regex=True)
```

```python
# dados['descricao_local'] = dados['descricao_local'].str.replace(' - ', ' ')
```

De maneira geral, o regex permite que os cientistas de dados processem, analisem e classifiquem grandes volumes de dados de texto de forma eficiente e automatizada. O uso adequado do regex pode ajudar a extrair informaÃ§Ãµes valiosas de dados de texto, alÃ©m de facilitar a limpeza e organizaÃ§Ã£o desses dados.
