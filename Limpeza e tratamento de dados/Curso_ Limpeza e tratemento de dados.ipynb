### Completando os dados

```python
idx = dados_normalizados[dados_normalizados['conta.cobranca.Total'] == ' '].index
idx_2 = dados_normalizados[(dados_normalizados['conta.cobranca.Total'] == ' ') & (dados_normalizados['conta.contrato'] == 'dois anos')].index
idx_1 = dados_normalizados[(dados_normalizados['conta.cobranca.Total'] == ' ') & (dados_normalizados['conta.contrato'] == 'um ano')].index
```

```python
dados_normalizados.loc[idx_2, 'cliente.tempo_servico'] = 24
dados_normalizados.loc[idx_1, 'cliente.tempo_servico'] = 12
```

```python
dados_normalizados.loc[idx, 'conta.cobranca.Total'] = dados_normalizados.loc[idx, 'conta.cobranca.mensal'] * dados_normalizados.loc[idx, 'cliente.tempo_servico']
```

```python
dados_normalizados.loc[idx, ['cliente.tempo_servico', 'conta.contrato', 'conta.cobranca.mensal', 'conta.cobranca.Total']]
```

O método `np.ceil()` serve apenas para arredondar os valores para cima.

### Retirando dados nulos

Primeiro vamos selecionar as colunas que usaremos como referência para excluir as linhas que possuem dados nulos.

```python
colunas_dropar = ['conta.contrato', 'conta.faturamente_eletronico', 'conta.metodo_pagamento']
```

Agora vamos verificar quantidade de linhas que possuem pelo menos um dado nulo dentro dessas 3 colunas.

```python
dados_completos[colunas_dropar].isnull().any(axis=1).sum()
```

Agora iremos remover essas linhas.

```python
dados_sem_nulo = dados_completos.dropna(subset=colunas_dropar).copy()
```

```python
dados_sem_nulo
```

```python
dados_sem_nulo.reset_index(inplace=True, drop=True)
```

```python
dados_sem_nulo
```

```python
dados_sem_nulo.isnull().sum()
```



> Agora você deve estar se perguntando, para onde foi os dados nulos das duas últimas colunas? Coincidentemente, as mesmas amostras que possuiam dados nulos nas 3 colunas anteriores, também tinham dados nulos nela. Portanto, nosso problema foi já foi resolvido.

> Outra pergunta que pode surgir é, por que não colocamos elas junto com as colunas de referência para a dropagem? Justamente por elas possuirem **dados relacionais** que poderia ser deduzidos a partir de outras colunas. Mas como não sobrou nenhum dado nulo, nosso trabalho foi concluído.

### Identificando Outliers

image.png

```python
import seaborn as sns
```

```python
sns.boxplot(x=dados_sem_nulo['cliente.tempo_servico'])
```

```python
Q1 = dados_sem_nulo['cliente.tempo_servico'].quantile(.25)
Q3 = dados_sem_nulo['cliente.tempo_servico'].quantile(.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5*IQR
limite_superior = Q3 + 1.5*IQR
```

```python
outliers_idx = (dados_sem_nulo['cliente.tempo_servico'] < limite_inferior) | (dados_sem_nulo['cliente.tempo_servico'] > limite_superior)
outliers_idx
```

```python
dados_sem_nulo[outliers_idx]['cliente.tempo_servico']
```

### Substituindo outliers

```python
dados_sem_out = dados_sem_nulo.copy()
```

```python
dados_sem_out.loc[outliers_idx, 'cliente.tempo_servico'] = np.ceil(
    dados_sem_out.loc[outliers_idx, 'conta.cobranca.Total'] / dados_sem_out.loc[outliers_idx, 'conta.cobranca.mensal']
)
```

```python
sns.boxplot(x=dados_sem_out['cliente.tempo_servico'])
```

```python
dados_sem_out[outliers_idx][['cliente.tempo_servico', 'conta.cobranca.mensal', 'conta.cobranca.Total']]
```

### Excluindo outliers

Agora precisamos refazer os cálculos, para identifiar os outliers que sobraram.

```python
Q1 = dados_sem_out['cliente.tempo_servico'].quantile(.25)
Q3 = dados_sem_out['cliente.tempo_servico'].quantile(.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5*IQR
limite_superior = Q3 + 1.5*IQR

outliers_idx = (dados_sem_out['cliente.tempo_servico'] < limite_inferior) | (dados_sem_out['cliente.tempo_servico'] > limite_superior)
outliers_idx
```

```python
dados_sem_out[outliers_idx]
```

```python
dados_sem_out = dados_sem_out[~outliers_idx]
```

```python
sns.boxplot(x=dados_sem_out['cliente.tempo_servico'])
```

```python
dados_sem_out.reset_index(drop=True, inplace=True)
```

```python
dados_sem_out
```

## Tratando variáveis Categóricas

### Removendo dados irrelevantes

Primeiramente iremos remover a coluna de `id_cliente`, pois ela não acrescenta nenhuma informação relevante para a análise de Machine Learning (ML).

```python
dados_sem_id = dados_sem_out.drop('id_cliente', axis=1).copy()
dados_sem_id
```

### Substituindo valores

Agora iremos 'numerizar' o maior número de colunas possíveis, pois os modelos de ML trabalham melhor com números. Para isso, podemos transformar as colunas com informações binárias (duas opções) para 0 e 1.

```python
mapeamento = {
    'nao': 0,
    'sim': 1,
    'masculino': 0,
    'feminino': 1,
}
```

```python
for col in dados_sem_id.columns:
  print('Coluna:', col)
  print(dados_sem_id[col].unique())
  print('-'*30)
```

```python
colunas_bi = ['telefone.servico_telefone', 'Churn', 'cliente.parceiro', 'cliente.dependentes', 'conta.faturamente_eletronico', 'cliente.genero']
```

```python
dados_sem_id[colunas_bi] = dados_sem_id[colunas_bi].replace(mapeamento)
```

```python
dados_sem_id
```

```python
for col in dados_sem_id.columns:
  print('Coluna:', col)
  print(dados_sem_id[col].unique())
  print('-'*30)
```
