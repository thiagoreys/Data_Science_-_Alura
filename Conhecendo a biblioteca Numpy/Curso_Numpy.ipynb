Agora podemos perceber que a diferença é bem menor, se tornando assim uma aproximação mais aceitável. Observe que, para este caso, a diferença nunca será próxima de zero, pois o gráfico oscila bastante.

O cálculo que fizemos foi por meio de aproximações intuitivas. Porém, existe uma fórmula para calcularmos os coeficientes angular e linear da reta, que nos permitem calcular o y e ajustar esta reta.


## Coeficiente angular (a)

formula.PNG

* â = Coeficiente angular;
* n = Número de elementos;
* Y = Cidade;
* X = Datas.

```python
Y = Moscow
X = datas
n = Moscow.size
```

```python
a = (n*sum(X*Y)-sum(X)*sum(Y))/(n*sum(X**2)-(sum(X))**2)
a
```

## Coeficiente linear (b)

image.png

```python
b = np.mean(Y) - a*np.mean(X)
```

Agora que temos a e b, podemos substituir na equação da reta:

```python
y = a*X + b
```

Obs: lembre-se que o Y em maiúsculo corresponde aos valores de Moscow, e neste cálculo queremos o valor de y (minúsculo).**negrito**



Vamos verificar a precisão do y utilizando a função `.linalg.norm()`:



```python
np.linalg.norm(Moscow-y)
```

Agora podemos replotar o gráfico com o novo valor de y:

```python
plt.plot(datas, Moscow)
plt.plot(x,y)
```

Perceba que não foi uma grande diferença se comparado com o resultado que obtemos anteriormente, mas temos um ajuste melhor. O processo que realizamos é uma regressão linear. Em seguida, entenderemos melhor para o que serve esta regressão.



### Motivo da regressão

A regressão linear serve para você estipular valores aproximados de números que não estão no array. Como por exemplo, podemos calcular a estimativa de preço na metade de um mês, como o 41.5, ou em um mês futuro, como o 100:

* y = 41.5a + b
* y = 100a + b

```python
plt.plot(datas,Moscow)
plt.plot(x,y)
plt.plot(41.5, 41.5*a + b,'*r')  # o *r serve para marcar com o asterisco o ponto do gráfico que estamos analisando.
plt.plot(100, 100*a + b,'*r')
```

## Gerando valores aleatórios em arrays

Valores aleatórios inteiros:

```python
np.random.randint(low=40, high=100, size=100)  # (min, max, quantidade)
```

Valores aleatórios floats:

```python
np.random.uniform(low=0.1, high=0.9, size=100)
```

Entretanto, esses valores na verdade são pseudoaleatórios, pois são baseados em um valor inicial semente que é utilizado para uma sequência de cálculos. Podemos fixar um valor inicial com a função `seed()`:

```python
np.random.seed(16)
np.random.uniform(low=0.1, high=0.9, size=100)
```

Não importa quantas vezes a função for gerada, os valores sempre serão os mesmos.

Agora iremos usar esses valores pseudoaletórios para gerar um array de coeficientes angulares:

```python
np.random.seed(84)
coef_angulares = np.random.uniform(low=0.10,high=0.90,size=100)
```

Com os coeficientes angulares armazenados na variável, calcularemos o valor da norma do gráfico de Moscow para cada um deles.

```python
normas = np.array([])
for i in range(100):
    normas = np.append(normas, np.linalg.norm(Moscow-(coef_angulares[i]*X+b)))
```

```python
normas
```

```python
# Menor valor da norma (o mais aproximado)
min(normas)
```

```python
# Index do menor valor
for i in range(len(normas)):
  if min(normas) == normas[i]:
    index = i
index
```

Como os valores das normas foram gerados de forma ordenada, se colocarmos esse index na lista de coeficientes angulares, veremos que sairá um coeficiente próximo daquele que encontramos utlizando a fórmula.

```python
coef_angulares[index]
```

Portanto, conseguimos encontrar o coeficiente angular da reta que mais aproxima da curva no gráfico, por meio de valores aleatórios.

### Seleção dos dados

Desses a colunas que iremos avaliar são as de diâmetro e peso. Crie um arrays específicos para guardar o diâmetro e peso da laranja e toranja. O diâmtro está na coluna zero e o peso na coluna 1. Os dados referentes a laranja vão até a linha 4999 e os referentes à toranja iniciam na linha 5000 do arquivo.

```python
diametro_laranja = dados[:5000,0]
diametro_toranja = dados[5000:,0]
peso_laranja = dados[:5000,1]
peso_toranja = dados[5000:,1]
```

### Visualização dos dados

```python
import matplotlib.pyplot as plt

plt.plot(diametro_laranja, peso_laranja)
plt.plot(diametro_toranja, peso_toranja)
plt.legend(['Laranja','Toranja'])
```

### O coeficiente das retas

As duas parecem ter uma inclinação muito parecida em uma primeira inspeção visual vamos determinar os coeficientes da reta para cada um dos casos.

Lembre que para o coeficiente angular precisamos implementar a função:
$$ \hat{a} = \frac{n*Soma( X_i*Y_i ) - Soma( X_i )*Soma(Y_i)}{n Soma(X_i^2) - (Soma(X_i))^2}$$

e para o coeficiente linear:
$$ \hat{b} = Media(Y_i) - \hat{a} * Media(X_i) $$

Perceba que nesse caso o X é o diâmtro e o Y é o peso. Além disso o n é o número de elementos dentro do X, por exemplo. Ele vai ser diferente para a laranja e toranja. Determine o coefiecente para cada caso:

#### Laranja

```python
Y = peso_laranja
X = diametro_laranja
n = np.size(X)
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2)-np.sum(X)**2)
a
```

```python
b = np.mean(Y) - a*np.mean(X)
b
```

#### Toranja

```python
Y = peso_toranja
X = diametro_toranja
n = np.size(X)
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2)-np.sum(X)**2)
a
```

```python
b = np.mean(Y) - a*np.mean(X)
b
```

Deu tudo certo, chegamos em valores diferentes para cada um dos casos. Então, embora a inclinação da reta seja parecida, descobrimos que os coeficientes da reta servem como uma forma de diferenciar se as frutas são laranjas ou toranjas.
