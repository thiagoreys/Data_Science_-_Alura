> Importante: Na aula nós realizamos a normalização em um arquivo com o formato JSON. No entanto, a função json_normalize() aceita apenas um dicionário ou uma lista de dicionários. Por isso, no vídeo foi usada a estratégia de usar o código `pd.json_normalize(dados_pacientes_2['Pacientes'])`, que retorna a lista de dicionários. Porém, se tentarmos usar parâmetros da função json_normalize em um arquivo JSON podem surgir erros. Para contornar isso, precisamos importar o módulo json e ler os arquivos conforme o código abaixo:



```python
import pandas as pd
import json

#Lendo o arquivo json usando o módulo Python JSON
with open('pacientes_2.json','r') as f:
    dados = json.loads(f.read())

#Normalizando os dados com os parâmetros record_path e meta
pd.json_normalize(dados, record_path='Pacientes', meta=['Pesquisa', 'Ano'])
```

### Para saber mais: obtendo arquivos JSON de APIs

“Uma empresa de alimentos saudáveis está enfrentando um problema com a popularidade de seus produtos. Apesar de oferecer opções saudáveis, muitos(as) clientes estão relutantes em comprar os produtos devido à falta de informação sobre os valores nutricionais das frutas que são utilizadas na produção dos alimentos”.

Nesta situação podemos acessar uma API chamada [Fruitvice](https://fruityvice.com) para coletar informações detalhadas sobre os valores nutricionais das frutas que são usadas nos produtos.

Uma maneira comum de acessar APIs é por meio da biblioteca [requests](https://https://pypi.org/project/requests/). Para isso, também é necessário importar o módulo `json`:

```python
import requests
import json
```

Os dados da API Fruitvice serão obtidos com a função `requests.get('url')`. Dentro dela nós passamos o endereço da API seguido de `api/fruit/all` para obter os dados de todas as frutas.

```python
dados_frutas = requests.get('https://fruityvice.com/api/fruit/all')
```

Conseguimos recuperar os resultados usando a função `json.loads()`. Para isso, é necessário passar a variável `dados_frutas`, criada anteriormente com o atributo `text` que retorna o conteúdo da resposta.

```python
resultado = json.loads(dados_frutas.text)
```

Em seguida, podemos visualizar o DataFrame:

```python
pd.DataFrame(resultado)
```

```python
pd.json_normalize(resultado)
```
